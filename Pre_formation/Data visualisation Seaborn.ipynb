```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# traitement réalisés précédemment
prets = pd.read_csv("D:\Dev\Mes fichiers Python\Fichiers de données\prets_final.csv")

prets.head()
```

```python
# on génère le nuage de point avec Seaborn
# on indique le dataframe, les colonnes du dataframe à placer en abscisse et ordonnées
import seaborn as sns
sns.scatterplot(data=prets, x='revenu', y='taux_endettement')
```

```python
# Avec l'argument 'hue', on sépare le résultat en fonction d'une autre colonne ici 'type'
sns.scatterplot(data=prets, x='revenu', y='taux_endettement', hue='type')
```

```python
# avec l'argument 'size', on change la taille du point en fonction d'une variable, ici la durée du prêt
sns.scatterplot(data=prets, x='revenu', y='taux_endettement', size='duree')
```

```python
plt.figure(figsize=(10,6))

# La fonctionrcParams.update , avec l’argumentfont.size , définit pour tous les graphiques – 
# et non pas uniquement le graphique actuel – 
# la taille de la police de tous les éléments (xlabel, title, legend, etc.) à 14.
plt.rcParams.update({'font.size': 14})

sns.scatterplot(data=prets, x='revenu', y='taux_endettement', hue='type')

plt.ylabel("Taux d'endettement")

plt.xlabel('Revenu mensuel (€)')

plt.grid()

plt.xlim(500, 7500)

# légende en dehors du graphique grâce à l'argument 'bbox_to_anchor'
plt.legend(bbox_to_anchor=(1, 1.02))

plt.title("Taux d'endettement en fonction du revenu mensuel\npar type de prêt contracté")

plt.show()
```

```python
# permet d'agréger les données directement dans le diagramme en barre, estimator correspond à la fonction d’agrégation
sns.barplot(data=prets, x='ville', y='remboursement', ci=None, estimator=sum)
```

```python
sns.barplot(data=prets, x='ville', y='remboursement', ci=None, estimator=np.mean, hue='type')
plt.figure(figsize=(10,6))
plt.rcParams.update({'font.size': 9})
plt.show()
```
