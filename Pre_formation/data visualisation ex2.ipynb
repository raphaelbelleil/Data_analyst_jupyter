```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

CA = pd.read_csv('https://raw.githubusercontent.com/benjaminmrl/data-4452741/main/CA.csv')
CA['date'] = pd.to_datetime(CA['date'])
CA.head()
```

```python
plt.plot(CA['date'], CA['immobilier'], color='blue', label = 'immobilier', linewidth=3)
plt.plot(CA['date'], CA['automobile'], color='orange', label = 'automobile', linewidth=3)
plt.plot(CA['date'], CA['consommation'], color='green', label = 'consommation', linewidth=3)
plt.title("Bénéfices mensuel sur l'année 2021, par type de prêt", fontsize = 14) # titre du graph
plt.ylabel("Bénéfices net (€)") # titre axe des ordonnée
plt.grid(color='gray', linewidth=0.5) # Quadrillage gris épaisseur trait de 0.5
plt.yticks(fontsize = 11) # graduation axe des ordonnées
plt.xticks(['2021-03','2021-06','2021-09','2021-12'],fontsize = 11) # graduation axe des abcisses
plt.legend(loc='upper right') # affichage légende en haut a droite 
plt.figure(figsize=(100, 70)) # taille du quadrillage 
```
