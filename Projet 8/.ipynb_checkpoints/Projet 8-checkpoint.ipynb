### Taux d'accès à l'eau potable en fonction du taux d'urbanisation

```python
# pivot population pour créer le taux de population urbaine par année par pays
population_pivot = population.pivot_table(index=['Country','Year'] ,columns='Granularity', values=['Year', 'Population'], dropna=False).reset_index()
population_pivot.head()
```

```python
# réarrangement des colonnes 
#population_pivot.columns = ['_'.join(x) for x in population_pivot.columns.values]
population_pivot.columns = ['Country', 'Year', 'Female', 'Male', 'Rural', 'Total', 'Urban']
```

```python
# création variables taux de chaque proportion de la population
population_pivot['female_rate'] = population_pivot['Female']/population_pivot['Total']
population_pivot['male_rate'] = population_pivot['Male']/population_pivot['Total']
population_pivot['rural_rate'] = population_pivot['Rural']/population_pivot['Total']
population_pivot['urban_rate'] = population_pivot['Urban']/population_pivot['Total']
```

```python
eau_potable_annee_pays = df[df['Granularity'] == 'Total'][['Country', 'Year', 'basic_service_water',
                                  'basic_service_water_population']].reset_index(drop=True)
```

```python
plt.figure(figsize=(15,9))
sns.scatterplot(data=population_pivot_2.groupby('Country').mean().reset_index(), x='urban_rate', y='basic_service_water')
```
