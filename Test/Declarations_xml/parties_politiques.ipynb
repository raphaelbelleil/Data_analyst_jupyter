## Consolidation des df

```python
def modif_nom(valeur):
    '''Enlève les accents et mets en majuscule.'''
    
    valeur = valeur.strip()
    # enlève les accents
    valeur = unidecode(valeur)
    valeur = valeur.upper()
    
    return valeur
```

```python
df = pd.concat([elus_conseiller_arrondissement, elus_conseiller_departemental, elus_depute, elus_epci, elus_maire,
           elus_membre_assemblees, elus_regional, elus_representant_parlement_europeen, elus_senateur], axis=0)
df.head()
```

```python
df.shape
```

```python
# modification nom, prenom, date

df['prenom_elu'] = df['prenom_elu'].apply(modif_nom)


df['date_naissance_elu'] = pd.to_datetime(df['date_naissance_elu'])


df['date_naissance'] = df['date_naissance_elu'].dt.strftime('%d') + '/' + df['date_naissance_elu'].dt.strftime('%m') + '/' + df['date_naissance_elu'].dt.strftime('%Y')
    
    
df = df.rename(columns={'nom_elu':'nom', 'prenom_elu':'prenom'})
```

## Jointure

```python
individus = pd.read_csv(folder + 'individus.csv')
individus
```

```python
# jointure entre individu et les tables parties
merge_partie = pd.merge(individus, df, on=['nom', 'prenom', 'date_naissance'], how='inner')
merge_partie.head(150)
```

```python
merge_partie.info()
```

```python
# jointure sans doublons, pour les individus présents dans 1 seul fichier
merge_partie_concat1 = merge_partie[~(merge_partie.duplicated(subset=['nom', 'prenom', 'date_naissance'], keep=False))]
merge_partie_concat1.head()
```

Pour les individus qui ont plusieurs parties politiques en fonction du fichier, on va associer le partie politique du fichier qui correspond au type mandat de la base de données initial.

```python
# individus présent dans plusieurs fichiers, potentiellement avec plusieurs parties politiques
merge_partie_avec_doublons = merge_partie[merge_partie.duplicated(subset=['nom', 'prenom', 'date_naissance'], keep=False)]
merge_partie_avec_doublons.head()
```

```python
def correspondance_tables(valeur):
    '''Renvoie le type mandat en fonction de la provenance de la table'''
    
    if valeur == 'elus_conseiller_departemental' : return 'Elu départemental'
    if valeur == 'elus_depute' : return 'Député'
    if valeur == 'elus_epci' : return 'Membre d’EPCI'
    if valeur == 'elus_maire' : return 'Maire ou adjoint municipal'
    if valeur == 'elus_regional' : return 'Elu régional'
    if valeur == 'elus_representant_parlement_europeen' : return 'Député européen'
    if valeur == 'elus_senateur' : return 'Sénateur'
    if valeur == 'elus_membre_assemblees' : return 'Collectivité à statut particulier'
    if valeur == 'elus_conseiller_arrondissement' : return 'Maire ou adjoint municipal'
```

```python
# colonne de correspondance
merge_partie_avec_doublons['correspondance_table_type_mandat'] = merge_partie_avec_doublons['table_provenance'].apply(correspondance_tables)
merge_partie_avec_doublons.head()
```

```python
# colonne qui renvoie 1 si la table de correspondance est égale au type mandat de l'autre base de données
merge_partie_avec_doublons['correspondance_table_type_mandat2'] = [1 if x==y else 0 for x,y in zip(merge_partie_avec_doublons['qualite_mandat_label_type_mandat'], merge_partie_avec_doublons['correspondance_table_type_mandat'])]
merge_partie_avec_doublons
```

```python
# individus avec le parti correspondant à sa fonction dans la 1ère bdd
merge_partie_concat2 = merge_partie_avec_doublons[merge_partie_avec_doublons['correspondance_table_type_mandat2']==1].drop(columns='correspondance_table_type_mandat2').drop_duplicates(subset=['nom', 'prenom', 'date_naissance'])
merge_partie_concat2.head()
```

```python
# id déja classés
id_individu_merge_partie_concat2 = merge_partie_concat2['id_individu'].values

# pour les id restants, dont on a pas fait la correspondance entre le type mandat et un des fichiers, on garde la 1ère ligne et le partie associé
merge_partie_concat3 = merge_partie_avec_doublons[~(merge_partie_avec_doublons['id_individu'].isin(id_individu_merge_partie_concat2))].drop_duplicates(subset=['nom', 'prenom',
'date_naissance']).drop(columns='correspondance_table_type_mandat2').drop_duplicates(subset=['nom', 'prenom', 'date_naissance'])
merge_partie_concat3.head()
```

```python
merge_partie_all = pd.concat([merge_partie_concat1, merge_partie_concat2,
                              merge_partie_concat3], axis=0).sort_values('nom').drop(columns='correspondance_table_type_mandat')
merge_partie_all.head()
```

```python
individus_sans_partie = individus[~individus['id_individu'].isin(merge_partie_all['id_individu'].values)]
individus_sans_partie.head()
```

```python
individus_sans_partie['qualite_mandat_label_type_mandat'].value_counts()
```

```python
# liaison avec les élus municipaux

date_naissance2 = []

for i in tqdm(range(elus_municipal.shape[0])):
    date2 = elus_municipal['date_naissance_elu'][i][8:10] + '/' + elus_municipal['date_naissance_elu'][i][5:7] + '/' + elus_municipal['date_naissance_elu'][i][:4]
    date_naissance2.append(date2)
    
        
elus_municipal['date_naissance'] = date_naissance2

# modification nom, prenom, date
elus_municipal['prenom_elu'] = elus_municipal['prenom_elu'].apply(modif_nom)
elus_municipal = elus_municipal.rename(columns={'nom_elu':'nom', 'prenom_elu':'prenom'})
```

```python
merge_partie_concat4 = pd.merge(individus_sans_partie, elus_municipal, on=['nom', 'prenom', 'date_naissance'], how='inner')
merge_partie_concat4.head()
```

```python
merge_partie_all2 = pd.concat([merge_partie_all, merge_partie_concat4], axis=0)
merge_partie_all2.head()
```

```python
merge_partie_all2.shape
```

```python
individus_sans_partie = individus[~individus['id_individu'].isin(merge_partie_all2['id_individu'].values)]
individus_sans_partie.head()
```

```python
merge_partie_all3 = merge_partie_all2.drop(columns=['date_naissance_elu', 'table_provenance'])
merge_partie_all3.head()
```
