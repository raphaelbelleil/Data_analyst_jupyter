## Nouveau script avec la récursivité

```python
tree = ET.parse(folder + 'extrait_declaration2.xml')

root = tree.getroot()
```

```python
# creation des listes de stockage
uuids2 = []
chemins2 = []
champss2 = []
valeurs2 = []
niveaux2 = []
```

```python
def parcourir_arbre(root, chemin="", niveau=0):
    # Obtenir le chemin complet
    chemin_actuel = f"{chemin}/{root.tag}" if chemin else root.tag
    

    # Si l'élément a du texte, ajoutez-le au tableau
    if root.text:
        
        chemins2.append(chemin_actuel)
        champss2.append(root.tag)
        valeurs2.append(root.text)
        niveaux2.append(niveau)
        
        # rang de la déclaration en cours sachant qu'on récupère le numéro de déclaration ('uuid') à chaque itération
        rang_uuid_actuel = (0 if champss2.count('declaration')<=1 else ((champss2.count('declaration')) - 1))
        # on prend la racine initial (tree.getroot()) car dans la fonction root est un paramètre
        uuid = tree.getroot()[rang_uuid_actuel].findall('uuid')[0].text
        uuids2.append(uuid)
        
        #print(f"Chemin: {chemin_actuel}, Balise: {root.tag}, Valeur: {root.text}, Niveau {niveau}, uuid {uuid}")
        

    # Parcourir les sous-éléments récursivement
    for enfant in root:
        parcourir_arbre(root=enfant, chemin=chemin_actuel, niveau = niveau+1)
```

```python
# Appelez la fonction pour parcourir l'arbre XML
parcourir_arbre(root)
```

```python
# code simple pour avoir le tag et le texte, quand il y a un texte
for element in root.iter() :
    if element.text:
        print(element.tag, element.text)
```

```python
def correspondance_types_postgresql(valeur):
    
    """ Retourne le format PostgreSQL en fonction du dtype de la colonne du Dataframe."""
    
    if valeur == 'object' : 
        return VARCHAR(150)
    
    if valeur == 'datetime64[ns]' : 
        return DATE()
    
    if valeur == 'int32' : 
        return INTEGER
    
    if valeur == 'int64' : 
        return INTEGER
    
    if valeur == 'float32' : 
        return FLOAT
    
    if valeur == 'float64' : 
        return FLOAT
```

```python
def types_postgresql(df):
    
    """Retourne un dictionnaire qui contient en clé le nom des colonnes du dataframe et en valeurs le type PostgreSQL."""
    
    champs_types_postgresql = df.dtypes.apply(correspondance_types_postgresql)
    
    dtype = {}

    for champs, types in zip(champs_types_postgresql.index, champs_types_postgresql.values):
        dtype[champs] = types

    return dtype
```
