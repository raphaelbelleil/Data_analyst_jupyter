```python
data3 = pd.read_csv(folder + 'data_nettoyees.csv')
```

```python
mask_colonne_table_individus = (data3['chemins'].str.contains('general/declarant/'))
```

```python
individus = data3[mask_colonne_table_individus]
```

```python
individus = individus[['id_declaration', 'champs', 'valeurs']]
individus
```

```python
# suppression des declarations/individus en doublons, suppression des lignes si pour une meme declaration plusieurs individus mentionnés
# on gardera les dernières valeurs
individus = individus.drop(individus[individus.duplicated(subset=['id_declaration', 'champs'], keep='last')].index)
```

```python
# pivot des champs
individus = pd.pivot(individus, columns='champs', index='id_declaration', values='valeurs').reset_index()
```

```python
individus['id_individu'] = individus['nom']+individus['prenom']+individus['dateNaissance']
individus
```

```python
individus.shape
```

```python
# suppression des doublons
individus = individus.drop_duplicates('id_individu').drop(columns='id_declaration').reset_index(drop=True)
individus.head()
```

```python
# encodage id_individus
encoder_id_individu = LabelEncoder()

individus['id_individu'] = encoder_id_individu.fit_transform(individus['id_individu'])
```
