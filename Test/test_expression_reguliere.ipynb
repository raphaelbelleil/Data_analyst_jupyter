```python
import re

pays_officiels = [
    'France', 'Allemagne', 'Italie', 'Espagne', 'Portugal', 'Royaume-Uni', 'États-Unis', 'Canada'
]

pays_fautes_orthographe = [
    'Franss', 'Alemagne', 'Italiee', 'Espania', 'Portugale', 'Roiaume-Unis', 'État-Unis', 'Canadda'
]

for pays_faute in pays_fautes_orthographe:
    correspondance_trouvee = False
    for pays_officiel in pays_officiels:
        pattern = fr"\b{re.escape(pays_officiel)}\b"
        if re.search(pattern, pays_faute, re.IGNORECASE):
            print(f'Correspondance trouvée : {pays_officiel} -> {pays_faute}')
            correspondance_trouvee = True
            break
    
    if not correspondance_trouvee:
        print(f'Aucune correspondance trouvée pour {pays_faute}')
```

```python
import difflib

pays_officiels = [
    'France', 'Allemagne', 'Italie', 'Espagne', 'Portugal', 'Royaume-Uni', 'États-Unis', 'Canada'
]

pays_fautes_orthographe = [
    'Franss', 'Alemagne', 'Italiee', 'Espania', 'Portugale', 'Roiaume-Unis', 'État-Unis', 'Canadda'
]

for pays_faute in pays_fautes_orthographe:
    correspondances = difflib.get_close_matches(pays_faute, pays_officiels, n=1, cutoff=0.6)
    if correspondances:
        correspondance = correspondances[0]
        print(f'Correspondance trouvée : {correspondance} -> {pays_faute}')
    else:
        print(f'Aucune correspondance trouvée pour {pays_faute}')
```
