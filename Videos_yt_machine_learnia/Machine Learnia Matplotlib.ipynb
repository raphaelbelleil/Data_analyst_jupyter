# Vidéo 14 MATPLOTLIB - Les Bases ! (14/30)

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

```python
x=np.linspace(0,10,11)
y=x**2
print(x)
print(y)
```

```python
plt.figure(figsize=(10,6))

plt.plot(x,y, c='red', lw=1, ls='--', label='x²')
plt.plot(x,x**3, label='x^3')

plt.title('Figure 1')
plt.xlabel('axe des x')
plt.ylabel('axe des y')

plt.legend()

plt.show()

plt.savefig('figure.png')
```

```python
plt.figure(figsize=(15,9))

plt.subplot(2,1,1)
plt.plot(x,y, label='x²')
plt.title('Figure 1')
plt.xlabel('axe des x')
plt.ylabel('axe des y')
plt.legend()

plt.subplot(2,1,2)
plt.plot(x,x**3, label='x^3', c='red')
plt.title('Figure 2')
plt.xlabel('axe des x')
plt.ylabel('axe des y')

plt.legend()

plt.show()
```

### Matplotlib orienté objet petit exemple

```python
# la figure et les axes sont créés comme des objets
# ax est un tableau numpy (2,1) dans ce cas 
fig, ax = plt.subplots(2,1, sharex=True)
ax[0].plot(x,y)
ax[1].plot(x, np.cos(x))
plt.show()
```

# Vidéo 15 MATPLOTLIB - Graphiques Importants (15/30)

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
# import du dataset des fleurs d'iris
iris = load_iris()

x = iris.data
y=iris.target
names = list(iris.target_names)

print (f'il y a {x.shape[0]} éléments, {x.shape[1]} variables et {np.unique(y).size} classes')
```

```python
def graphique(data):
    n=len(data)
    plt.figure(figsize=(12,8))
    for k, i in zip(data.keys(), range(1, n+1)):
        plt.subplot(n, 1, i)
        plt.plot(data[k])
        plt.title(k)
```

```python
print ('variable x : variables attributs des iris : longueur sépale, largeur sépale, longueur pétale, largeur pétale')
print(x)
```

### Plt scatter

```python
plt.figure(figsize=(10,6))

# c=y --> on modifie la couleur en fonction de la classe, pour chaque valeur unique de y on met une nouvelle couleur
# ici [1,2,3]--> 3 valeurs de y donc 3 couleurs
# alpha : transparence,  s : taille, on peut prendre une variable pour la taille
plt.scatter(x[:,0], x[:,1], c=y, alpha=0.5, s=x[:,3]*100)
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')

# s=x[:,3]*100 --> longeur des pétales
plt.show()
```

### Graphique 3D 2

```python
f = lambda x,y : np.sin(x) + np.cos(x+y)
```

```python
X = np.linspace(0, 5, 100)
Y = np.linspace(0, 5, 100)
X, Y = np.meshgrid(X,Y)

Z=f(X, Y)
```

```python
Z.shape
```

```python
ax= plt.axes(projection='3d')
ax.plot_surface(X,Y,Z, cmap='plasma')
```

### Histogramme

```python
# import du dataset des fleurs d'iris
iris = load_iris()

x = iris.data
y=iris.target
names = list(iris.target_names)

print (f'il y a {x.shape[0]} éléments, {x.shape[1]} variables et {np.unique(y).size} classes')
```

```python
# nombre de catégorie qu'on va trouver dans l'histogramme
plt.hist(x[:,0], bins=20)
plt.hist(x[:,1], bins=20)
```

```python
plt.figure(figsize=(10,6))
plt.hist2d(x[:,0], x[:,1], cmap='Blues')
plt.xlabel('longueur sépal')
plt.ylabel('largeur sépal')
plt.colorbar()
plt.show()
```

```python
# distribution des couleurs d'une image --> en fonction des coleurs de chaque pixel
from scipy import misc
face = misc.face(gray=True)
#plt.imshow(face, cmap='gray')
# l'image est transformé en vecteur pour analyser sa distribution
plt.hist(face.ravel(), bins=255)
plt.show()
```

### Contour plots
