### Regrouper les données par mois

### Evolution du bitcoin en faisant la moyenne pour chaque mois

```python
btc.loc['2019', 'Close'].resample('M').mean().plot(figsize=(10,6))
```

### En faisant la moyenne pour 2 semaines 

```python
btc.loc['2019', 'Close'].resample('2W').mean().plot(figsize=(10,6))
```

```python
# on utilise resample sur 2 semaines et sur l'écart type --> donne une idée sur la volatilité du bitcoin
btc.loc['2019', 'Close'].resample('2W').std().plot(figsize=(10,6))
```

```python
plt.figure(figsize=(12, 8))
btc.loc['2019', 'Close'].plot()
btc.loc['2019', 'Close'].resample('M').mean().plot(label='moyenne par mois', lw=3, ls=':', alpha=0.8)
btc.loc['2019', 'Close'].resample('W').mean().plot(label='moyenne par semaine', lw=2, ls='--', alpha=0.8)
plt.legend()
plt.show()
```

### Agregate

```python
# affiche l'évolution de la moyenne et l'écart type de Close par mois
btc.loc['2019','Close'].resample('M').agg(['mean', 'std']).plot()
```

```python
m = btc['Close'].resample('W').agg(['mean', 'std', 'min', 'max'])

plt.figure(figsize=(12, 8))
m['mean']['2019'].plot(label='moyenne par semaine')
# permet de créer cet halo bleu qui va entourer la courbe du min et du max de Close pour chaque valeur de l'index ici chaque semaine
plt.fill_between(m.index, m['max'], m['min'], alpha=0.2, label='min-max par semaine')

plt.legend()
plt.show()
```

### Moyenne mobile, moving average

```python
plt.figure(figsize=(10,6))

btc.loc['2019', 'Close'].plot(label='Evolution du BTC')

btc.loc['2019', 'Close'].resample('W').mean().plot(label='Evolution du BTC, moyenne par semaine')

btc.loc['2019', 'Close'].rolling(window=7).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile sur 7 jours')

plt.legend()
```

```python
plt.figure(figsize=(10,6))

btc.loc['2019-09', 'Close'].plot(label='Evolution du BTC')

btc.loc['2019-09', 'Close'].resample('W').mean().plot(label='Evolution du BTC, moyenne par semaine')

# ici on utilise center=True pour placer le point correspondant à la moyenne au milieu des 7 jours
btc.loc['2019-09', 'Close'].rolling(window=7, center=True).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile sur 7 jours')

plt.legend()
```

```python
plt.figure(figsize=(10,6))

btc.loc['2019-09', 'Close'].plot(label='Evolution du BTC')

btc.loc['2019-09', 'Close'].rolling(window=7).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile sur 7 jours')

# ici on utilise center=True pour placer le point correspondant à la moyenne au milieu des 7 jours
btc.loc['2019-09', 'Close'].rolling(window=7, center=True).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile sur 7 jours, center=True')

plt.legend()
```

### Moyenne mobile exponentielle

```python
plt.figure(figsize=(10,6))

btc.loc['2019-09', 'Close'].plot(label='Evolution du BTC')

btc.loc['2019-09', 'Close'].rolling(window=7).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile sur 7 jours')

# ici on utilise center=True pour placer le point correspondant à la moyenne au milieu des 7 jours
btc.loc['2019-09', 'Close'].rolling(window=7, center=True).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile sur 7 jours, center=True')

btc.loc['2019-09', 'Close'].ewm(alpha=0.6).mean().plot(ls='--', lw=2, alpha=0.6, label='Evolution du BTC, moyenne mobile exponentielle')

plt.legend()
```

### Exercice

```python
plt.figure(figsize=(10,6))

btc.loc['2019', 'Close'].plot(label='close')

btc.loc['2019', 'Close'].rolling(window=28).min().plot(label='min')

btc.loc['2019', 'Close'].rolling(window=28).max().plot(label='max')

plt.legend()
```

```python
data = btc.copy()
data['Buy'] = np.zeros(len(data))
data['Sell'] = np.zeros(len(data))
```

```python
data['RollingMax'] = data['Close'].shift(1).rolling(window=28).max()
data['RollingMin'] = data['Close'].shift(1).rolling(window=28).min()
data.loc[data['RollingMax'] < data['Close'], 'Buy'] = 1
data.loc[data['RollingMin'] > data['Close'], 'Sell'] = -1
```

```python
start ='2019'
end='2019'
fig, ax = plt.subplots(2, figsize=(12, 8), sharex=True)
#plt.figure(figsize=(12, 8))
#plt.subplot(211)
ax[0].plot(data['Close'][start:end])
ax[0].plot(data['RollingMin'][start:end])
ax[0].plot(data['RollingMax'][start:end])
ax[0].legend(['close', 'min', 'max'])
ax[1].plot(data['Buy'][start:end], c='g')
ax[1].plot(data['Sell'][start:end], c='r')
ax[1].legend(['buy', 'sell'])
```
