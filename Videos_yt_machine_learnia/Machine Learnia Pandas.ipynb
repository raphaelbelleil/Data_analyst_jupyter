## Vidéo 18 PANDAS PYTHON Tutoriel Français - Time Series (18/30)

```python
# on met la date en index et parse_dates va modifier le type de l'index 
btc=pd.read_csv('BTC-EUR.csv', index_col='Date', parse_dates=True)
```

```python
btc.head()
```

```python
btc.shape
```

```python
btc['Close'].plot(figsize=(10,6))
```

```python
# l'index a changé de type --> DatetimeIndex
btc.index
```

```python
btc.loc['2012':'2014', 'Close'].plot()
```

### Regrouper les données par mois

```python
btc.loc['2019', 'Close'].resample('M').plot(figsize=(10,6))
```

### Evolution du bitcoin en faisant la moyenne pour chaque mois

```python
btc.loc['2019', 'Close'].resample('M').mean().plot(figsize=(10,6))
```

### En faisant la moyenne pour 2 semaines 

```python
btc.loc['2019', 'Close'].resample('2W').mean().plot(figsize=(10,6))
```

```python
btc.loc['2019', 'Close'].resample('2W').std().plot(figsize=(10,6))
```

```python
btc.loc['2019', 'Close'].resample('M').plot(figsize=(10,6))
btc.loc['2019', 'Close'].resample('M').mean().plot(figsize=(10,6))
btc.loc['2019', 'Close'].resample('2W').mean().plot(figsize=(10,6))
btc.loc['2019', 'Close'].resample('2W').std().plot(figsize=(10,6))

plt.show()
```
