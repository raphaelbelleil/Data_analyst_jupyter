```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as st
from scipy.stats import norm
import statsmodels as stm
```

```python
guerison = pd.read_csv('guerison.txt')
guerison.head()
```

```python
guerison.value_counts()
```

```python
taux_guerison_emprique=guerison.value_counts()[1]/guerison.shape[0]
taux_guerison_emprique
```

```python
# estimateur ponctuel --> ma proportion 
taux_guerison_emprique_pourcent = round(100*guerison.value_counts()[1]/guerison.shape[0],2)
taux_guerison_emprique_pourcent
```

## Remarque

Par la suite, on part du principe que notre échantillon suit une loi normale (tests paramétriques), on devrait normalement effectuer des tests de normalités au préalable, ou une loi de Bernouilli (quel test dans ce cas ?)

## Calcul de l'interval de confiance

```python
n=guerison.shape[0]
alpha = 0.05
borne_inf = taux_guerison_emprique-(norm.ppf(1-alpha/2))*(np.sqrt(taux_guerison_emprique*(1-taux_guerison_emprique)/n))
borne_sup = taux_guerison_emprique+(norm.ppf(1-alpha/2))*(np.sqrt(taux_guerison_emprique*(1-taux_guerison_emprique)/n))
```

```python
intervalle_confiance = [round(100*borne_inf,2), round(100*borne_sup,2)]
```

```python
print(f"L'interval de confiance de l'échantillon est {intervalle_confiance}")
```

## Autre méthode

```python
# Interval de confiance avec statsmodel, en fonction de notre estimateur, n, et le risque alpha, on utilise la méthode normal car l'échntillon est grand (>30)
from statsmodels.stats.proportion import proportion_confint
proportion_confint(guerison.value_counts()[1],n, alpha=alpha, method='normal')
```

## Test statistique
Sur une proportion

```python
# test statistique sur une proportion --> test binomial
n_gueri = guerison.value_counts()[1]
stm.stats.proportion.binom_test(n_gueri, n, prop = 0.75, alternative = 'larger')
```

On a une pvalue = 0.24, très supérieur aux seuils usuels (0.01, 0.05), cela signifie qu'on accepterait H0 avec 24% de chance de se tromper. Donc, on rejette ici H0 et on accepte H1

--> p>p0, on met sur le marché le nouveau médicament.
