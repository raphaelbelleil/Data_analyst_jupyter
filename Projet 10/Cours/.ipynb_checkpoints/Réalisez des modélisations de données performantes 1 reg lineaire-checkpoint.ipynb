# Régression linéaire multiple

```python
y = ozone['maxO3']
X = ozone.drop(columns=['obs', 'maxO3', 'vent', 'pluie'])
```

```python
X = sm.add_constant(X)
model = sm.OLS(y, X).fit()

model.summary()
```

## Significativité des paramètres

On fait passer un test de Student à chaque variable explicative (par rapport à la variable cible) pour vérifier qu'elle est utile au modèle, qu'elle est significative.

Le test de Student vérifie l'indépendance ou non de 2 échantillons.  

Ici on veut que la variable cible soit dépendante de la variable explicative.  
On enlève donc les variables explicatives qui n'ont pas d'impact sur la variable cible.

```python
import functions as f
```

```python
# fonction qui teste un à un le modèle, vérifie la significativité des variables et retire les variables non significatives
# intégrer le df avec la variable cible + la variable cible
model = f.backward_selected(ozone.drop(columns=['obs', 'vent', 'pluie']), 'maxO3')
model
```

Le modèle final est composé de Vx9 + Ne9 + T12 + maxO3v en variables explicatives et de maxO3 en variable cible.

```python
model.summary()
```

## Atypicité de certaines données et influence

### Levier

A chaque observation est associée une valeur de levier, c'est à dire un indice d'atypicité.

Une observation est considérée trop atypique si elle dépasse un seuil défini par 2*p/n avec p nombre de variables explicatives (+1 avec la colonne de biais) et n nombre d'observations.

```python
ozone2 = ozone[['maxO3', 'Vx9', 'Ne9', 'T12', 'maxO3v']]
ozone2['levier'] = model.get_influence().hat_matrix_diag
ozone2.head()
```

```python
ozone2.shape
```

```python
# nombre de variables
p = model.params.shape[0]

# nombre d'observations
n = ozone2.shape[0]

# seuil levier
seuil_levier = 2*p/n

ozone2[ozone2['levier']>seuil_levier]
```

```python
# représenation graphique
plt.bar(ozone2.index, ozone2['levier'])
plt.plot(np.linspace(0, ozone2.shape[0], 2), [seuil_levier, seuil_levier], c='r')
```

### Résidus standardisés

Autre moyen d'identifier les valeurs atypiques.

```python
ozone2['rstudent'] = model.get_influence().resid_studentized_internal
ozone2
```

```python
# seuil pour les résidus standardisés
alpha = 0.05
seuil_rstudent  = t.ppf(1-alpha/2,n-p-1)
```

```python
# représenation graphique
plt.bar(ozone2.index, ozone2['rstudent'])
plt.plot(np.linspace(0, ozone2.shape[0], 2), [seuil_rstudent, seuil_rstudent], c='r')
plt.plot(np.linspace(0, ozone2.shape[0], 2), [-seuil_rstudent, -seuil_rstudent], c='r')
```

### Distance de Cook

```python
ozone2['distance_cook'] = model.get_influence().summary_frame()['cooks_d']
seuil_dcook = 4/(n-p)
ozone2
```

```python
# représenation graphique
plt.bar(ozone2.index, ozone2['distance_cook'])
plt.plot(np.linspace(0, ozone2.shape[0], 2), [seuil_dcook, seuil_dcook], c='r')
```

```python
ozone2[ozone2['distance_cook']>seuil_dcook]
```

## Colinéarité des variables
Calcul des VIF
Vérifier que les variables explicatives ne s'influence pas entre elles.

```python
# donnees passées dans le modèle
variables = model.model.exog
[variance_inflation_factor(variables, i) for i in np.arange(1,variables.shape[1])]
```

Ici, tous les coefficients sont inférieurs à 10, il n'y a donc pas de problème de colinéarité.

## Hétéroscédasticité des résidus
Constance de la variance des résidus

```python
_, pval, __, f_pval = het_breuschpagan(model.resid, variables)
print('p value test Breusch Pagan:', pval)
```

La p-valeur ici n'est pas inférieure à $5$%, on ne rejette pas l'hypothèse $H0$ selon laquelle les variances sont constantes (l'hypothèse d’homoscédasticité).
