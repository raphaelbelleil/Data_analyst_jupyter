# Les Arbres de décision

Arbres de décision :
- Facile à comprendre et à interpréter.
- Peut gérer des données catégorielles et numériques.
- Peut souffrir de surajustement (overfitting), mais cela peut être géré avec des techniques comme la taille de l'arbre limitée (élagage).

## Fonctionnement 

- Sélection de la caractéristique (feature) la plus importante : L'algorithme commence par choisir la caractéristique la plus pertinente à partir de l'ensemble de données pour diviser les données en sous-groupes. La sélection de cette caractéristique se fait en fonction de critères comme l'entropie, l'indice de Gini ou le gain d'information, qui évaluent l'importance de la caractéristique pour la tâche de classification ou de régression.
****
- Division des données : Une fois la caractéristique choisie, les données sont divisées en sous-ensembles en fonction de ses valeurs. Par exemple, si la caractéristique est "Âge" et que la valeur de division est de 30 ans, les données seront divisées en deux groupes : les individus de moins de 30 ans et ceux de 30 ans ou plus. On détermine une fois de plus la caractéristique et la valeur de division grâce à l'indice de gini et au coût du niveau.
****

- Répétition du processus : Les deux sous-ensembles nouvellement créés sont ensuite soumis au même processus de sélection de caractéristique et de division, créant ainsi des nœuds enfants. Ce processus se répète récursivement jusqu'à ce qu'une condition d'arrêt soit atteinte, telle qu'un nombre maximum de niveaux, un nombre minimum d'échantillons dans un nœud ou l'homogénéité des données dans le nœud (c'est-à-dire que toutes les données appartiennent à la même classe dans le cas de la classification).
****

- Attribution d'une classe ou d'une valeur : Une fois l'arbre construit, chaque feuille (nœud terminal) est associée à une classe (dans le cas de la classification) ou à une valeur (dans le cas de la régression). Lorsqu'un nouvel exemple est présenté à l'arbre, il suit les règles de division basées sur les caractéristiques pour arriver à une feuille, où il est classé ou une valeur est prédite.

## Notes

- Classification et Régression
- Données Linéaire ou non linéaire
- Fonctionne bien pour les données catégorielles
- Nécessite peu de préparation, pas de normalisation (standardisation, log...) ni de mise à l'échelle
