## Interval de confiance

### Sur la moyenne

```python
x=essence['conso']
alpha = 0.01
n=len(x)
int_inf=x.mean()-(st.t.ppf(1-alpha/2,n-1))*x.std(ddof=1)/np.sqrt(n)
int_sup=x.mean()+(st.t.ppf(1-alpha/2,n-1))*x.std(ddof=1)/np.sqrt(n)
```

```python
intervalle_confiance = [round(int_inf,2), round(int_sup,2)]
print(f"L'interval de confiance de l'échantillon est {intervalle_confiance}")
```

### Sur la variance

```python
int_inf2 = (n-1)*x.var(ddof=0)/st.chi2.ppf(1-alpha/2,n-1)
int_sup2 = (n-1)*x.var(ddof=0)/st.chi2.ppf(alpha/2,n-1)
```

```python
intervalle_confiance = [round(int_inf2,2), round(int_sup2,2)]
print(f"L'interval de confiance de l'échantillon est {intervalle_confiance}")
```
