# Modèle avec sklearn

```python
var = data.drop(columns=target).columns
```

```python
# colonnes categ
var_categ = list(data.select_dtypes(include='object').columns)

# data categ 2
data_categ = data.select_dtypes(include='object')
data_categ
```

```python
# catégories
categories = []
for col in data_categ:
    for element in data_categ[col].unique():
        categories.append(element)
categories
```

```python
# colonnes num
var_num = list(data.drop(columns=target).select_dtypes(exclude='object').columns)

# data num 2
data_num = data.drop(columns=target).select_dtypes(exclude='object')
data_num
```

```python
# création du train et test set
data_train, data_test = train_test_split(data, test_size=0.2)

print(f"Les dimensions du train_set sont {data_train.shape}")
print(f"Les dimensions du test_set sont {data_test.shape}")
```

```python
# creation preprocessor
categ_pipeline = make_pipeline(OneHotEncoder())
#pipeline_selection = make_pipeline(SelectKBest(f_regression, k=5))
#num_pipeline = make_pipeline(PolynomialFeatures(2))
    
preprocessor = make_column_transformer((categ_pipeline, var_categ),                                    
                                       remainder='passthrough')

preprocessor
```

```python
def X_y(data, target):
    
    y = data[[target]]
    X = data.drop(columns=target)
    
    return X, y
```

```python
# creation des jeux d'entrainement et de test
X_train, y_train = X_y(data_train, target)
X_test, y_test = X_y(data_test, target)
```

```python
# creation de l'estimateur
reg = LinearRegression()

model = make_pipeline(preprocessor, reg)
model
```

```python
# entrainement sur le jeu d'entrainement
model.fit(X_train, y_train)
```

```python
cross_val_score(model, X_train, y_train, cv=5)
```

```python
model.score(X_test, y_test)
```

```python
# courbe d'apprentissage
warnings.simplefilter('ignore')

N, train_score, val_score = learning_curve(model, X_train, y_train, train_sizes=np.linspace(0.1, 1, 10), cv=5)

plt.title("Courbe d'apprentissage")
plt.plot(N, train_score.mean(axis=1), label='train_set')
plt.plot(N, val_score.mean(axis=1), label='val_set')
plt.xlabel("Quantité de données")
plt.ylabel("R²")
plt.ylim((0,1.2))
plt.legend()
plt.show()
```

# Modèle avec stat model

```python
data.head()
```

```python
# encodage des variables qualitatives

encoder = OneHotEncoder(sparse_output=False)

df_categ_encoder = pd.DataFrame(encoder.fit_transform(data[var_categ]), columns=categories, index=data.index)

df_categ_encoder.head()
```

```python
# df données pour modèle avec stat model
data_sm = pd.concat((data[target], data[var_num], df_categ_encoder), axis=1)
data_sm.head()
```

```python
# création du train et test set
data_train, data_test = train_test_split(data_sm, test_size=0.2)

# division X et y
X_train, y_train = X_y(data_train, target)
X_test, y_test = X_y(data_test, target)

# creation de la liste des colonnes
columns = data_sm.drop(columns=target).columns
```

```python
# ajout de la colonne de biais
df_X_train = pd.DataFrame(X_train, columns=columns)
df_X_test = pd.DataFrame(X_test, columns=columns)
X_train_sm = sm.add_constant(df_X_train)
X_test_sm = sm.add_constant(df_X_test)

# creation du modèle
results = sm.OLS(y_train, X_train_sm).fit()

results.summary()
```

```python
# R² effectué sur le jeu de test

y_pred = results.predict(X_test_sm)

r_squared = r2_score(y_test, y_pred)

r_squared
```

```python
# calcul des VIF
vif = [variance_inflation_factor(X_train_sm, i) for i in range(X_train_sm.shape[1])]
df_vif = pd.DataFrame(vif, index=X_train_sm.columns, columns=['VIF'])
df_vif
```

VIF infinie car variables binaires égale à 0.

```python
scores = []

for iteration in range(10):
    # création du train et test set
    data_train, data_test = train_test_split(data_sm, test_size=0.2)

    # division X et y
    X_train, y_train = X_y(data_train, target)
    X_test, y_test = X_y(data_test, target)

    # ajout de la colonne de biais
    df_X_train = pd.DataFrame(X_train, columns=columns)
    df_X_test = pd.DataFrame(X_test, columns=columns)
    X_train_sm = sm.add_constant(df_X_train)
    X_test_sm = sm.add_constant(df_X_test)

    # creation du modèle
    results = sm.OLS(y_train, X_train_sm).fit()

    y_pred = results.predict(X_test_sm)

    r_squared = r2_score(y_test, y_pred)
    
    scores.append(r_squared)

print("R² en fonction des itérations :")    
print(scores)
print("Moyenne R²:", np.mean(scores))
print("Médiane R²:", np.median(scores))
print("Ecart type:", np.std(scores))
```

```python
# stabilité du modèle

plt.plot(range(len(scores)), scores)
plt.title("Evolution du R² au cours des itérations")
plt.xlabel("Itérations")
plt.ylabel("R²")
plt.ylim(0, max(scores)*1.2)
```

```python
# normalité des résidus

print(results.resid.max())

sns.histplot(results.resid)

shapiro(results.resid)
```
