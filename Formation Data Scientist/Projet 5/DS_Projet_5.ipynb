```python
# renvoie une liste des mots pr√©sents dans la description
def tokenisation(valeur):
    
    # mettre en minuscule
    valeur = valeur.lower()
    
    # tokeniser, separer chaque mot
    valeur = nltk.word_tokenize(valeur)

    # enlever les caract√®res unidecode, par exemple les accents
    valeur = [unidecode(element) for element in valeur]

    # enlever les caract√®res speciaux de la liste de token
    valeur = [enlever_caracteres_speciaux(element) for element in valeur]

    # enlever les mots plus petit que 3 caract√®res
    valeur = [element for element in valeur if len(element)>2]
    
    return valeur
```

```python
def nombre_mots(valeur):
    return len(valeur)
```

```python
def nombre_mots_unique(valeur):
    return len((pd.Series(valeur)).unique())
```

```python
def occurence_mots(valeur):
    return dict(pd.Series(valeur).value_counts())
```

```python
df['description_token'] = df['description'].apply(tokenisation)
```

```python
df['nombre_mots'] = df['description_token'].apply(nombre_mots)
```

```python
df['nombre_mots_unique'] = df['description_token'].apply(nombre_mots_unique)
```

```python
df['occurence_mots'] = df['description_token'].apply(occurence_mots)
```
