## Bases

L'objectif est de transformer les données pour transformer un problème non linéaire en problème linéaire.

On transforme l'espace initial en un espace de redescription pour représenter différemment les données.

Espace de Hilbert produit scalaire de 2 espaces.

```python
X, y = make_circles(n_samples=400, factor=.3, noise=.05)
```

```python
plt.scatter(X[:,0], X[:,1], c=y)
```

On ne peut pas séparer ces données par une droite, on voudrait les séparer par un cercle.

de la forme : 

x1² + x2² = R²
x1² + x2² - R² = 0

fi1 = x1² et fi2 = x2²

fi1 + fi2 - R² = 0 qui est l'équation d'une droite


En faisant cette transformation, on transforme le problème en un problème linéaire.

En résumé

    Il est possible d’appliquer un algorithme linéaire à un problème non linéaire.

    Une façon de construire une fonction de décision non linéaire est de redécrire les données dans un nouvel espace dans lequel on apprend une fonction de décision linéaire.

## SVM à noyau

Les noyaux permettent de traiter des problèmes non linéaire de manière linéaire.

Les noyaux permettent de projeter les données dans un espace de caractéristiques de dimension supérieure (espace de Hilbert), où une séparation linéaire est possible. Les noyaux couramment utilisés incluent le noyau linéaire, le noyau polynomial et le noyau gaussien (RBF).

Synthèse : 
- séparer des données non linéaire
- créer un espace de retranscription (de plus grande dimensions) pour passer d'un probleme non lineaire à lineaire grace aux noyaux
- les noyaux sont des produits scalaires d'espaces
- les noyaux peuvent être lineaire, polynomial ou gaussien (rbf)...

### Gamma

A l'instantiation d'une svm, gamma va déterminer la portée d'influence de chaque observation et donc de son influence sur la frontière de décision.

si gamma petit, importance des observations sur une plus grande region de l'espace --> risque de sous apprentissage car ne capte pas forcement les détails des données, la fonction noyau est plus globale

si gamma grand, importance des observations sur une plus petite region de l'espace --> risque de sur apprentissage car la fonction noyau est très locale et s'adapte précisement aux données

### C

C'est la tolérance aux erreurs de classification, plus précisement le compromis entre minimisation des erreurs de classification et maximisation de la marge.

C faible, marge maximiser, plus d'erreurs, modèle simple, risque de sous apprentissage

C élevé, erreur minimisé, marge plus petite, modèle plus complexe, risque de sur apprentissage

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import confusion_matrix
```

```python
data = pd.read_csv("winequality-white.csv", delimiter=';')
data.head()
```

```python
X = data.drop(columns='quality')
y = data['quality']
```

```python
X_scaler = StandardScaler().fit_transform(X)
y_bin = np.where(y<6, 0, 1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_scaler, y_bin, test_size=0.2, random_state=4)
```

```python
# Modèle linéaire pour avoir une base

reg = LogisticRegression(max_iter=10000)

reg.fit(X_train, y_train)

reg.score(X_test, y_test)
```

```python
# svm à noyau

ksvm = SVC(kernel='rbf', degree=3, gamma='scale', random_state=4)

ksvm.fit(X_train, y_train)

ksvm.score(X_test, y_test)
```

```python
# prédire sur le jeu de test
y_pred = ksvm.decision_function(X_test)
# distance par rapport à la frontière de decision

# construire la courbe ROC
from sklearn import metrics
fpr, tpr, thr = metrics.roc_curve(y_test, y_pred)

# calculer l'aire sous la courbe ROC
auc = metrics.auc(fpr, tpr)

# créer une figure
from matplotlib import pyplot as plt
fig = plt.figure(figsize=(6, 6))

# afficher la courbe ROC
plt.plot(fpr, tpr, '-', lw=2, label='gamma=0.01, AUC=%.2f' % auc)

# donner un titre aux axes et au graphique
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('SVM ROC Curve', fontsize=16)

# afficher la légende
plt.legend(loc="lower right", fontsize=14)

# afficher l'image
plt.show()
```

```python
# grid search cv

# choisir 6 valeurs pour C, entre 1e-2 et 1e3
C_range = np.logspace(-2, 3, 6)

# choisir 4 valeurs pour gamma, entre 1e-2 et 10
gamma_range = np.logspace(-2, 1, 4)

# grille de paramètres
param_grid = {'C': C_range, 'gamma': gamma_range}


grid = GridSearchCV(ksvm, param_grid, cv=5, scoring='roc_auc')

grid.fit(X_train, y_train)
```

```python
grid.best_params_
```

```python
grid.best_score_
```

```python
best_params = grid.best_params_
```

```python
# svm à noyau

ksvm2 = SVC(**best_params)

ksvm2.fit(X_train, y_train)

ksvm2.score(X_test, y_test)
```

```python
# prédire sur le jeu de test avec le modèle optimisé
y_test_pred_cv = ksvm2.decision_function(X_test)

# construire la courbe ROC du modèle optimisé
fpr_cv, tpr_cv, thr_cv = metrics.roc_curve(y_test, y_test_pred_cv)

# calculer l'aire sous la courbe ROC du modèle optimisé
auc_cv = metrics.auc(fpr_cv, tpr_cv)

# créer une figure
fig = plt.figure(figsize=(6, 6))

# afficher la courbe ROC précédente
plt.plot(fpr, tpr, '-', lw=2, label='gamma=0.01, AUC=%.2f' % auc)

# afficher la courbe ROC du modèle optimisé
plt.plot(fpr_cv, tpr_cv, '-', lw=2, label='gamma=%.1e, AUC=%.2f' % \
         (grid.best_params_['gamma'], auc_cv))
         

# donner un titre aux axes et au graphique
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('SVM ROC Curve', fontsize=16)

# afficher la légende
plt.legend(loc="lower right", fontsize=14)

# afficher l'image
plt.show()
```

```python
# score sur le jeu de test

svc.score(X_test, y_test)
```

```python
# distance à la frontière de decision

y_pred_prob = svc.decision_function(X_test)
```

```python
# roc curve et calcul de l'auc

fpr, tpr, thr = metrics.roc_curve(y_test, y_pred_prob)

# calculer l'aire sous la courbe ROC
auc = metrics.auc(fpr, tpr)

# afficher la courbe ROC
plt.plot(fpr, tpr, '-', lw=2, label='gamma=50, AUC=%.2f' % auc)

# donner un titre aux axes et au graphique
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('SVM ROC Curve', fontsize=16)

# afficher la légende
plt.legend(loc="lower right", fontsize=14)

# afficher l'image
plt.show()
```

```python
param_grid = {
    'C' : np.logspace(-2, 2, 5),
    'gamma' : np.logspace(-2, 2, 5)
}

svm = SVC()

grid = GridSearchCV(svm, param_grid, cv=5, scoring='roc_auc')

grid.fit(X_train, y_train)
```

```python
best_params = grid.best_params_
best_params
```

```python
# meilleurs resultats du modèle
pd.DataFrame(grid.cv_results_).sort_values('rank_test_score').head()
```

```python
# selection du meilleur estimateur

svc_fin = grid.best_estimator_
```

```python
# prédire sur le jeu de test avec le modèle optimisé
y_test_pred_cv = svc_fin.decision_function(X_test)

# construire la courbe ROC du modèle optimisé
fpr_cv, tpr_cv, thr_cv = metrics.roc_curve(y_test, y_test_pred_cv)

# calculer l'aire sous la courbe ROC du modèle optimisé
auc_cv = metrics.auc(fpr_cv, tpr_cv)

# créer une figure
fig = plt.figure(figsize=(6, 6))

# afficher la courbe ROC précédente
plt.plot(fpr, tpr, '-', lw=2, label='gamma=50, AUC=%.2f' % auc)

# afficher la courbe ROC du modèle optimisé
plt.plot(fpr_cv, tpr_cv, '-', lw=2, label='gamma=%.1e, AUC=%.2f' % \
         (grid.best_params_['gamma'], auc_cv))
         

# donner un titre aux axes et au graphique
plt.xlabel('False Positive Rate', fontsize=16)
plt.ylabel('True Positive Rate', fontsize=16)
plt.title('SVM ROC Curve', fontsize=16)

# afficher la légende
plt.legend(loc="lower right", fontsize=14)

# afficher l'image
plt.show()
```

# Régression avec la SVM, Kernel Régression Ridge

Méthode non linéaire de régression

```python
from sklearn.kernel_ridge import KernelRidge
from sklearn.metrics import mean_squared_error
```

```python
data.head()
```

```python
X = data.drop(columns='quality')
y = data['quality']

X_scaler = StandardScaler().fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X_scaler, y, test_size=0.2, random_state=4)
```

```python
kridge = KernelRidge( alpha=1.0, # valeur par défaut 
    kernel='rbf', # noyau Gaussien
    gamma=0.01)

kridge.fit(X_train, y_train)

y_pred = kridge.predict(X_test)

rmse = np.sqrt(mean_squared_error(y_test, y_pred))
rmse
```

```python
r2 = kridge.score(X_test, y_test)
r2
```

```python
# créer une figure
fig = plt.figure(figsize=(6, 6))

# Compter, pour chaque paire de valeurs (y, y') où y est un vrai score et y' le score prédit,
# le nombre de ces paires.
# Ce nombre sera utilisé pour modifier la taille des marqueurs correspondants 
# dans un nuage de points
sizes = {}
for (yt, yp) in zip(list(y_test), list(y_pred)):
    if (yt, yp) in sizes.keys():
        sizes[(yt, yp)] += 1
    else:
        sizes[(yt, yp)] = 1        
keys = sizes.keys()

# afficher les prédictions
plt.scatter([k[0] for k in keys], 
            [k[1] for k in keys], 
             s=[sizes[k] for k in keys], 
            label="gamma = 0.01: RMSE = {:.2f}".format(rmse))

# étiqueter les axes et le graphique
plt.xlabel('Vrai score', fontsize=16)
plt.ylabel(u'Score prédit', fontsize=16)
plt.title('kernel Ridge Regression', fontsize=16)

# limites des axes
plt.xlim([2.9, 9.1])
plt.ylim([2.9, 9.1])

# afficher la légende
plt.legend(loc="lower right", fontsize=12)
```

```python
# valeurs du paramètre C
alpha_range = np.logspace(-2, 2, 5)

# valeurs du paramètre gamma
gamma_range = np.logspace(-2, 1, 4)

# grille de paramètres
param_grid = {'alpha': alpha_range, 'gamma': gamma_range}

# score pour sélectionner le modèle optimal
score = 'neg_mean_squared_error'

# initialiser la validation croisée
grid_pred = GridSearchCV(
        KernelRidge(kernel='rbf'),
        param_grid,
        cv=5,
        scoring=score)
                                    
# exécuter la validation croisée sur le jeu d'entraînement
grid_pred.fit(X_train, y_train)

# prédire sur le jeu de test avec le modèle sélectionné 
y_test_pred_cv = grid_pred.predict(X_test)

# calculer la RMSE correspondante
rmse_cv = np.sqrt(metrics.mean_squared_error(y_test, y_test_pred_cv))
```
