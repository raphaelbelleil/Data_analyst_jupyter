```python
# Import des librairies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
```

```python
import keras
from keras import layers
import tensorflow as tf
```

```python
df = pd.read_csv('ecg.csv', names=[f'col_{i}' for i in range(pd.read_csv('ecg.csv').shape[1])])
df = df.rename(columns={'col_140':'anomalie'})
df.head()
```

```python
# preparation des donnees

X = StandardScaler().fit_transform(df.drop(columns='anomalie'))
y = df['anomalie']

# creation jeu de train et de test

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
# initialisation du modèle
model = tf.keras.Sequential()

# dimensions données d'entrée
model.add(keras.Input(shape=(X.shape[1])))

# creation de la première couche
model.add(layers.Dense(140, input_dim=2, activation='sigmoid'))

# creation de la 2e couche
model.add(layers.Dense(70, activation='sigmoid'))

# creation de la couche de sortie
model.add(layers.Dense(1, activation='sigmoid'))
```

```python
# compilation du modèle

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs=50, batch_size=4, verbose=1)
```

```python
model.evaluate(X_test, y_test, verbose=1)
```
